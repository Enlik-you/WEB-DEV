```python
def minn(a, b, c, d):
    arr = [a, b, c, d]
    minnn = a
    for i in range(4):
        if arr[i] < minnn:
            minnn = arr[i]
    return minnn

a, b, c, d = map(int, input().split())

print(minn(a, b, c, d))
```

```python
def power(a, b):
    return a ** b

a, n = input().split()
a = float(a)
b = int(n)

print(power(a, b))
```

```python
def xor(a, b):
    if a == b:
        return 0
    else:
        return 1

a, b = map(int, input().split())
print(xor(a, b))
```
